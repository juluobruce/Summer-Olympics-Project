#  Python -- Data Cleaning

```python
#import tools and libraries:
import pandas as pd
import csv
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
#open and read csv files and store into a data frame:

events = "Resources/athlete_events.csv"
regions = "Resources/noc_regions.csv"

events_pd = pd.read_csv(events)
regions_pd = pd.read_csv(regions)


```

```python
#Structrue for events:
events_pd.head()
```

```python
#structure for regoins
regions_pd.head()
```

```python
#merge tables together 
olympics = pd.merge(events_pd, regions_pd, on=['NOC'])
```

```python
#new table structure 
olympics.head(10)
```

```python
#clean up data 
olympics = olympics.drop(['NOC','notes','Height','Weight'], axis=1)

#olympics.head()

```

```python
olympics.head()
```

```python
# Converting csv file to download as data 
olympics.to_csv('olympics.csv', index=False)
```

```python
#checking the types
#olympics.dtypes
#olympics.info()
#olympics['region'].unique()
#olympics['region'].value_counts() 

#focusing on summer track 
#olympics['Sport'] = olympics['Sport'].replace({'Athletics': 'Track & Field'})
track_olympics = olympics.loc[olympics['Sport'] == 'Track & Field']



```

```python
#Top 5 particiapting countries (USA,UK,Germany,Russia,France)

#seperate each regions to be able to do the graph 
us = track_olympics.loc[olympics['region'] == 'USA']
germany = track_olympics.loc[olympics['region'] == 'Germany']
france = track_olympics.loc[olympics['region'] == 'France']
uk = track_olympics.loc[olympics['region'] == 'UK']
russia = track_olympics.loc[olympics['region'] == 'Russia']
```

```python
#clean up 'Athletics to Track & Field
us['Sport'] = us['Sport'].replace({'Athletics': 'Track & Field'})
germany['Sport'] = germany['Sport'].replace({'Athletics': 'Track & Field'})
france['Sport'] = france['Sport'].replace({'Athletics': 'Track & Field'})
uk['Sport'] = uk['Sport'].replace({'Athletics': 'Track & Field'})
russia['Sport'] = russia['Sport'].replace({'Athletics': 'Track & Field'})

```

## Bubble plots

```python
#US
plt.figure(figsize=(10,10))
plt.scatter(
        x=us[us['Medal'] == 'Gold']['Age'],
        y=us[us['Medal'] == 'Gold']['Year'],
        c = '#FFD700',
        s = us[us['Medal'] == 'Gold'].count(),
        alpha = 0.75,
        label = 'Gold',
        edgecolors="black",
        linewidths=0.60)

plt.scatter(
        x=us[us['Medal'] == 'Silver']['Age'],
        y=us[us['Medal'] == 'Silver']['Year'],
        c= '#bcc3c6',
        s = us[us['Medal'] == 'Silver'].count(),
        alpha = 0.75,
        label = 'Silver',
        edgecolors="black",
        linewidths=0.60)

plt.scatter(
        x=us[us['Medal'] == 'Bronze']['Age'],
        y=us[us['Medal'] == 'Bronze']['Year'],
        c= '#FF7F50',
        s = us[us['Medal'] == 'Bronze'].count(),
        alpha = 0.75,
        label = 'Bronze',
        edgecolors="black",
        linewidths=0.60)

plt.legend( title="Medal type")
plt.xlabel("Age")
plt.ylabel("Year")
plt.title("USA Track & Field Medal data")
plt.show()
```
